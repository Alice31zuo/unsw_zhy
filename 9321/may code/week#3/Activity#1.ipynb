# Activity#1

```python
import pandas as pd
```

```python
#read csv file 
df= pd.read_csv('Books.csv')
dl= pd.read_csv('City.csv')
```

```python
#read the dataframe

print(",".join([column for column in df]))

for index,row in df.iterrows():
      print(",".join([str(row[column]) for column in df]))
```

```python
drop_columns=['Edition Statement',
                       'Corporate Author',
                       'Corporate Contributors',
                       'Former owner',
                       'Engraver',
                       'Contributors',
                       'Issuance type',
                       'Shelfmarks']
```

```python
# drop columns 
df.drop(drop_columns,inplace=True,axis=1)
df
```

```python
#The percentage of NaN in the data per column
num_row= df.shape[0]
for column in df:
    present=100*df[column].isnull().sum()/num_row
    print(column,present)
```

# Activity#2

```python
# modified some values 
df['Place of Publication']
df['Place of Publication']=df['Place of Publication'].apply(lambda x:'London'if 'London'in x else x.replace('-',''))
df['Place of Publication']
```

```python
#extract a specific data 
data=df['Date of Publication'].str.extract(r'(\d{4})',expand=False)
#Convert "Date of Publication" cells to numbers
data=pd.to_numeric(data)
    
```

```python
#Replace NaN with 0 for the cells of "Date of Publication"
df['Date of Publication']=data.fillna(0)
```

```python
df
```

# Activity#3

```python
#Replace the spaces in the column names with the underline character ('_') 
df.columns=[c.replace(' ','_') for c in df.columns]
```

```python
df
```

```python
#Filter the rows and only keep books which are published in "London" after 1866.
df.query('Date_of_Publication > 1866 and Place_of_Publication=="London"')
```

```python
df
```

# Activity#4

```python
#Merge two datasets based on the name of city
#how=Use keys from left frame only

combine_datasets=pd.merge(df,dl,how='right',left_on='Place_of_Publication',right_on='City')
```

```python
combine_datasets
```

```python
#Group by the resultant dataframe based on the country column 
gb_df = combine_datasets.groupby(['Country'],as_index=False)
```

```python
#Use count() to calculate the number of publications by country.
df = gb_df['Identifier'].count()
```
