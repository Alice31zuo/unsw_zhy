```python
import pandas as pd
import sqlite3
from  pandas.io import sql
```

```python
#read csv file 
df=pd.read_csv('Demographic_Statistics_By_Zip_Code.csv')
```

```python
# Store the dataframe in sqlite databas
connection=sqlite3.connect('Demographic_Statistics.db')
sql.to_sql(df, name='Demographic_Statistics', con=connection)
```

```python
#Query the database and load the data into a new dataframe again
sql.read_sql('select * from ' +'Demographic_Statistics', connection)
```
