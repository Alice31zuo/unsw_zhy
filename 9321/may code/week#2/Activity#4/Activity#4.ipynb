```python
import requests
import pandas as pd
```

```python
#create a get request and fetch the JSON 
resp = requests.get(url='https://data.cityofnewyork.us/api/views/kku6-nxdu/rows.json')
json_obj = resp.json()


```

```python
#Create a dataframe out of the JSON object
json_data = json_obj['data']
columns = []
for c in json_obj['meta']['view']['columns']:
        columns.append(c['name'])

pd.DataFrame(data=json_data, columns=columns)
```
